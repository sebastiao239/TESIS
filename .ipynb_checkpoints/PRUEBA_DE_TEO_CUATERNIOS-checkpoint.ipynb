```julia
using LinearAlgebra  #####Son las paqueterias que uso, para instalar una paqueteria basta colocar Pkg.add("Nombre_de_la_paqueteria")
using Plots
using SpecialFunctions: gamma
using Random, Distributions
using StatsBase
using Alert
```

```julia
function OUH(x0::Float64,γ::Float64,t::AbstractArray,Δt::Float64,η::Float64,cte::Float64=0.0) #Proceso de Ornstein–Uhlenbeck
    x=Vector{Float64}(undef, length(t)) #Definimos un vector lleno de basura para luego sobre escribir en el 
    x[1]=x0    #La primera entrada es la condicion inicial del Proceso de Ornstein–Uhlenbeck
    
    for i=1:length(t)                 #Usamos un for para poder hacer el proceso
        x0=x0-γ*Δt*(x0-cte)+(η*sqrt(Δt)*randn()) # Es el proceso de Ornstein–Uhlenbeck que depende del paso anterior 
        @inbounds x[1+i]=x0                 #@inbounds ignora si la lista tiene el tamaño adecuado y sobreescribimos en la lista basura
    end
    return x                          # Devolvemos la caminata de Ornstein–Uhlenbeck
end
```

```julia
function eigenR(n::Int64,Δt::Float64,K::Array{Float64,1},T::Array{Float64,1}) 
    R=diagm(0 =>ones(3)) #Hacemos una matriz llena de unos en la diagonal de dimension 2
    for i=1:n            #Hacemos un for donde n es la cantidad de pasos de t, pasos de tiempo
       R=(diagm(0 => ones(3))+diagm(1 => [K[i],T[i]]*Δt,-1 => -[K[i],T[i]]*Δt)+(0.5*diagm(1 => [K[i],T[i]]*Δt,-1 => -[K[i],T[i]]*Δt)^2))*R #Hacemos los pasos de multiplicar la matriz por la matriz anterior 
    end
    R #Obtenemos el eigenvalor y eigenvector
end
```

```julia
k0=3.0
t0=3.0

γK=2.0
γT=2.0

ηK=1.0
ηT=1.0

N=10.0

Δt=0.001

t=0:Δt:N

ctek=3.0
ctet=3.0

#para animar
tiradas=10000
step=300
length(1:step:length(t))
length(t)
```

```julia
K=OUH(k0,γK,t,Δt,ηK,ctek)
T=OUH(t0,γT,t,Δt,ηT,ctet)
plot(t,K,legend=false)
plot!(t,T,legend=false)
```

# VAMOS A GENERAR LO QUE NECESITO! i.e. la triada o vector

$$S=\left(\frac{\tau}{\sqrt{\kappa^2+\tau^2}},\frac{\kappa}{\sqrt{\kappa^2+\tau^2}},\Delta t^{\prime}\sqrt{\kappa^2+\tau^2} \right)=(u_{x},u_{z},\phi) $$

Las primeras entradas del vector $S$ son las componentes del eje de rotación de una matriz de rotación $R_n$ i.e. 

$$u_n=(u_n^x,0,u_n^z)=\left(\frac{\tau_n}{\sqrt{\kappa_n^2+\tau_n^2}},0,\frac{\kappa_n}{\sqrt{\kappa_n^2+\tau_n^2}}\right)$$

La última componente del vector $S$ no es otra cosa que el angulo $\phi_n$ de una matriz de rotación $R_n$ i.e.

$$\phi_n=\Delta t^{\prime}\sqrt{\kappa_n^2+\tau_n^2}$$

Con estas ecuaciones podemos generar lo necesario para generar las multiplicaciones de matrices que queremos, i.e. las $A_n$

$$A_n=\prod_{i=0}^n R_i$$

```julia
U(K::Float64,T::Float64)=[T/sqrt(K^2+T^2),0,K/sqrt(K^2+T^2)]
ϕ(K::Float64,T::Float64,Δt::Float64)=Δt*sqrt(K^2+T^2)
```

```julia
@time begin
    Un=U.(K,T)
    ϕn=ϕ.(K,T,Δt)
end;
```

# Primer paso

Se necesita calcular estos valores

$$
\psi_1 =2arcos\left[\left(1-\frac{\phi_1^2}{8}\right) \left(1-\frac{\phi_0^2}{8}\right)-\frac{\phi_1}{2}\frac{\phi_0}{2}(u_1\cdot u_0) \right]\\
w_1^x= \left[\left(1-\frac{\phi_1^2}{8}\right)\frac{\phi_0}{2}u_0^x + \left(1-\frac{\phi_0^2}{8}\right)\frac{\phi_1}{2}u_1^x \right]csc\left( \frac{\psi_1}{2}\right)\\
w_1^y=\left[ \left(u_1^z u_0^x-u_0^z u_1^x \right)\frac{\phi_1}{2}\frac{\phi_0}{2}  \right]csc\left( \frac{\psi_1}{2}\right) \\
w_1^z=\left[\left(1-\frac{\phi_1^2}{8}\right)\frac{\phi_0}{2}u_0^z + \left(1-\frac{\phi_0^2}{8}\right)\frac{\phi_1}{2}u_1^z \right]csc\left( \frac{\psi_1}{2}\right)
$$

```julia
cos_2(x)=1-x^2/8

ψ0(ϕ0,ϕ1,u,w)=2*acos(cos_2(ϕ1)*cos_2(ϕ0)-(ϕ1/2)*(ϕ0/2)*dot(u,w))

w1x(ϕ0,ϕ1,u,w,ψ0)=(cos_2(ϕ1)*(ϕ0/2)*w[1]+cos_2(ϕ0)*(ϕ1/2)*u[1])*csc(ψ0/2)

w1y(ϕ0,ϕ1,u,w,ψ0)=((u[3]*w[1]-w[3]*u[1])*(ϕ1/2)*(ϕ0/2))*csc(ψ0/2)

w1z(ϕ0,ϕ1,u,w,ψ0)=(cos_2(ϕ1)*(ϕ0/2)*w[3]+cos_2(ϕ0)*(ϕ1/2)*u[3])*csc(ψ0/2)
```

```julia
ψ1=ψ0(ϕn[1],ϕn[2],Un[1],Un[2])
```

# Segundo y n-Pasos

Se necesita calcular estos valores

$$
\psi_n =2arcos\left[\left(1-\frac{\phi_n^2}{8}\right) cos\left(\frac{\psi_{n-1}}{2}\right)-\frac{\phi_n}{2}sin\left(\frac{\psi_{n-1}}{2}\right)(u_n\cdot w_{n-1}) \right]\\
w_n^x= \left[\left(1-\frac{\phi_n^2}{8}\right)sin\left(\frac{\psi_{n-1}}{2}\right)w_{n-1}^x + cos\left(\frac{\psi_{n-1}}{2}\right)\frac{\phi_n}{2}u_n^x - \left(w_{n-1}^y u_n^z\right)sin\left(\frac{\psi_{n-1}}{2}\right)\frac{\phi_n}{2}  \right]csc\left( \frac{\psi_n}{2}\right)\\
w_n^y=\left[\left(1-\frac{\phi_n^2}{8}\right)sin\left(\frac{\psi_{n-1}}{2}\right)w_{n-1}^y + \left(u_n^z w_{n-1}^x-w_{n-1}^z u_n^x \right)sin\left(\frac{\psi_{n-1}}{2}\right)\frac{\phi_n}{2}  \right]csc\left( \frac{\psi_n}{2}\right) \\
w_n^z=\left[\left(1-\frac{\phi_n^2}{8}\right)sin\left(\frac{\psi_{n-1}}{2}\right)w_{n-1}^z + cos\left(\frac{\psi_{n-1}}{2}\right)\frac{\phi_n}{2}u_n^z + \left(w_{n-1}^y u_n^x\right)sin\left(\frac{\psi_{n-1}}{2}\right)\frac{\phi_n}{2}  \right]csc\left( \frac{\psi_n}{2}\right)
$$

# Construir lo que queremos
Ya sabemos como es el vector unitario que queremos $w_n$ y el ángulo de rotación $\psi_n$ para cada $A_n$ que tiene la forma

$$
\begin{bmatrix}
\cos (\psi_n) +(w_n^x)^2 \left(1-\cos (\psi_n)\right) & w_n^x w_n^y \left(1-\cos (\psi_n)\right) - w_n^z \sin (\psi_n) & w_n^x w_n^z \left(1-\cos (\psi_n)\right) + w_n^y \sin (\psi_n) \\ w_n^y w_n^x \left(1-\cos (\psi_n)\right) + w_n^z \sin (\psi_n) & \cos (\psi_n) + (w_n^y)^2\left(1-\cos (\psi_n)\right) & w_n^y w_n^z \left(1-\cos (\psi_n)\right) - w_n^x \sin (\psi_n) \\ w_n^z w_n^x \left(1-\cos (\psi_n)\right) - w_n^y \sin (\psi_n) & w_n^z w_n^y \left(1-\cos (\psi_n)\right) + w_n^x \sin (\psi_n) & \cos (\psi_n) + (w_n^z)^2\left(1-\cos (\psi_n)\right)
\end{bmatrix}
$$

```julia
function R0(Δt::Float64,K::Array{Float64,1},T::Array{Float64,1},i::Int64) 
    R=(diagm(0 => ones(3))+diagm(1 => [K[i],T[i]]*Δt,-1 => -[K[i],T[i]]*Δt)+(0.5*diagm(1 => [K[i],T[i]]*Δt,-1 => -[K[i],T[i]]*Δt)^2)) #Hacemos los pasos de multiplicar la matriz por la matriz anterior 
    return R
end
```
